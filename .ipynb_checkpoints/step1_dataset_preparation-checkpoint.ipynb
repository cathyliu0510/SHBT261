## Goal:
- Work with the Caltech-101 dataset (101 object cate-
gories, ∼9,000 images).

- Train and evaluate at least three different methods for image classification. For example
ResNet, EfficientNet, and one classical ML.

- Compare classical machine learning and deep learning methods.

### Step 1: Dataset Preparation
Shared pipeline for all three methods: ResNet, EfficientNet, and Classical ML.

**Dataset**: [Caltech-101 on Kaggle](https://www.kaggle.com/datasets/imbikramsaha/caltech-101)  
1. Seeds & imports — PyTorch, torchvision, sklearn, PIL all seeded at SEED=42 for full reproducibility.
2. Configuration — One place to set DATA_ROOT, IMG_SIZE (default 224, changeable to 64/128 for ablations), batch size, and split ratios.
3. Dataset scan — Walks the class folders, excludes BACKGROUND_Google (a known junk class in Caltech-101), builds image_paths and image_labels lists.
4. Class imbalance analysis — Plots the distribution. Caltech-101 is notably imbalanced (airplanes ~800 images, some classes ~40). Outputs class_distribution.png.
5. Stratified split — Two-stage train_test_split giving exact 70/15/15 with class proportions preserved in all three splits.
6. Class weights — compute_class_weight on train labels only, class_weights_tensor ready to pass to nn.CrossEntropyLoss().
7. Transform pipelines — Three variants:

- train_transform — with augmentation (flip, rotation, color jitter, random crop)
- eval_transform — deterministic resize + normalize
- train_transform_no_aug — for ablation study

8. Custom Dataset + DataLoaders — Caltech101Dataset handles greyscale/RGBA edge cases via .convert("RGB").
9. Sanity check — Visualizes 16 augmented training images with class labels.
10. Export — Saves shared_data.pkl containing all split paths, labels, class names, and config. All three model will load this file to guarantee identical splits.

```python
# import packages
import random
import shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from collections import Counter

# PyTorch
import torch
import torchvision
from torchvision import transforms, datasets
from torch.utils.data import DataLoader, Subset

# Sklearn
from sklearn.model_selection import train_test_split

# Set seed
SEED = 261
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
# Ensures deterministic behavior in cuDNN
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

print(f"PyTorch: {torch.__version__} | CUDA available: {torch.cuda.is_available()}")
```

```python
# Configuration
# UPDATE this path to wherever you extracted the Kaggle download
DATA_ROOT = Path("/Users/yichengliu/Desktop/mini-project1/caltech-101")

# Image size config — used in ablation studies (try 64 or 128)
IMG_SIZE = 224  # default for pretrained ResNet/EfficientNet

# Split ratios
TRAIN_RATIO = 0.70
VAL_RATIO   = 0.15
TEST_RATIO  = 0.15

BATCH_SIZE  = 32
NUM_WORKERS = 0

# Verify dataset root exists
assert DATA_ROOT.exists(), f"Dataset not found at: {DATA_ROOT}\nPlease update DATA_ROOT."
print(f"Dataset root found: {DATA_ROOT}")
```

```python
# ── Dataset exploration ────────────────────────────────────────────────────────
# Collect all image paths and their class labels
image_paths, image_labels = [], []

# Caltech-101 includes a junk 'BACKGROUND_Google' folder — we exclude it
EXCLUDE_CLASSES = {"BACKGROUND_Google"}

class_dirs = sorted([d for d in DATA_ROOT.iterdir()
                     if d.is_dir() and d.name not in EXCLUDE_CLASSES])
class_names = [d.name for d in class_dirs]
class_to_idx = {name: i for i, name in enumerate(class_names)}

for cls_dir in class_dirs:
    for img_path in cls_dir.iterdir():
        if img_path.suffix.lower() in {".jpg", ".jpeg", ".png"}:
            image_paths.append(str(img_path))
            image_labels.append(class_to_idx[cls_dir.name])

print(f"Total images : {len(image_paths)}")
print(f"Total classes: {len(class_names)}")
print(f"Classes sample: {class_names[:5]} ... {class_names[-3:]}")
```

```python
# ── Class imbalance check ──────────────────────────────────────────────────────
label_counts = Counter(image_labels)
counts_series = pd.Series({class_names[k]: v for k, v in sorted(label_counts.items(), key=lambda x: -x[1])})

print(f"Min images per class: {counts_series.min()} ({counts_series.idxmin()})")
print(f"Max images per class: {counts_series.max()} ({counts_series.idxmax()})")
print(f"Mean images per class: {counts_series.mean():.1f}")

# Visualize class distribution — top 30 and bottom 30 for readability
fig, axes = plt.subplots(1, 2, figsize=(18, 5))
counts_series.head(30).plot(kind="bar", ax=axes[0], color="steelblue")
axes[0].set_title("Top 30 classes (most samples)")
axes[0].set_ylabel("Image count")
axes[0].tick_params(axis='x', rotation=90)

counts_series.tail(30).plot(kind="bar", ax=axes[1], color="salmon")
axes[1].set_title("Bottom 30 classes (fewest samples)")
axes[1].tick_params(axis='x', rotation=90)
plt.tight_layout()
plt.savefig("class_distribution.png", dpi=100, bbox_inches="tight")
plt.show()
print("\nClass imbalance is present — handled via stratified split + class-weighted loss during training.")
```

```python
# ── Stratified train / val / test split ───────────────────────────────────────
# Step 1: Split off test set (15%), stratified by label
train_val_paths, test_paths, train_val_labels, test_labels = train_test_split(
    image_paths, image_labels,
    test_size=TEST_RATIO,
    stratify=image_labels,
    random_state=SEED
)

# Step 2: Split remaining 85% into train (70/85) and val (15/85)
val_fraction = VAL_RATIO / (TRAIN_RATIO + VAL_RATIO)
train_paths, val_paths, train_labels, val_labels = train_test_split(
    train_val_paths, train_val_labels,
    test_size=val_fraction,
    stratify=train_val_labels,
    random_state=SEED
)

print(f"Train  : {len(train_paths)} images ({len(train_paths)/len(image_paths)*100:.1f}%)")
print(f"Val    : {len(val_paths)} images ({len(val_paths)/len(image_paths)*100:.1f}%)")
print(f"Test   : {len(test_paths)} images ({len(test_paths)/len(image_paths)*100:.1f}%)")
```

```python
# ── Compute class weights for imbalance handling
# Used as `weight` param in nn.CrossEntropyLoss during training
from sklearn.utils.class_weight import compute_class_weight

class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.arange(len(class_names)),
    y=train_labels
)
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float32)

print(f"Class weight range: [{class_weights.min():.3f}, {class_weights.max():.3f}]")
print("class_weights_tensor is ready for nn.CrossEntropyLoss(weight=class_weights_tensor)")
```

```python
# ── Transform pipelines 
# These ImageNet normalization stats are used for all pretrained deep learning models
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]

# --- Training transform: with augmentation (ablation: compare to no_aug below) ---
train_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),  # resize to target size
    transforms.RandomHorizontalFlip(p=0.5),  # random flip
    transforms.RandomRotation(degrees=15),  # random rotation +- 15 degree
    transforms.ColorJitter(
        brightness=0.3, contrast=0.3,
        saturation=0.2, hue=0.1),  # color variation
    transforms.RandomResizedCrop(
        IMG_SIZE, scale=(0.8, 1.0)),  # random crop + zoom
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# --- Val/Test transform: deterministic, no augmentation ---
eval_transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

# --- Ablation: no-augmentation training transform ---
train_transform_no_aug = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

print("Transforms defined: train_transform | eval_transform | train_transform_no_aug")
```

```python
# ── Custom Dataset class 
from torch.utils.data import Dataset
from PIL import Image

class Caltech101Dataset(Dataset):
    """Lightweight dataset wrapping pre-split path/label lists."""

    def __init__(self, paths, labels, transform=None):
        self.paths     = paths
        self.labels    = labels
        self.transform = transform

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        # Open image; convert to RGB to handle greyscale/RGBA edge cases
        img = Image.open(self.paths[idx]).convert("RGB")
        if self.transform:
            img = self.transform(img)
        return img, self.labels[idx]


# ── Instantiate datasets
train_dataset = Caltech101Dataset(train_paths, train_labels, transform=train_transform)
val_dataset   = Caltech101Dataset(val_paths,   val_labels,   transform=eval_transform)
test_dataset  = Caltech101Dataset(test_paths,  test_labels,  transform=eval_transform)

# ── DataLoaders
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE,
                          shuffle=True,  num_workers=NUM_WORKERS, pin_memory=False)
val_loader   = DataLoader(val_dataset,   batch_size=BATCH_SIZE,
                          shuffle=False, num_workers=NUM_WORKERS, pin_memory=False)
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE,
                          shuffle=False, num_workers=NUM_WORKERS, pin_memory=False)

print(f"Train batches: {len(train_loader)} | Val batches: {len(val_loader)} | Test batches: {len(test_loader)}")
```

```python
# ── Sanity check: visualize a batch
def denormalize(tensor, mean=IMAGENET_MEAN, std=IMAGENET_STD):
    """Reverse ImageNet normalization for visualization."""
    mean = torch.tensor(mean).view(3, 1, 1)
    std  = torch.tensor(std).view(3, 1, 1)
    return (tensor * std + mean).clamp(0, 1)

images, labels = next(iter(train_loader))
fig, axes = plt.subplots(2, 8, figsize=(20, 6))
for i, ax in enumerate(axes.flat):
    if i >= len(images): break
    img_np = denormalize(images[i]).permute(1, 2, 0).numpy()
    ax.imshow(img_np)
    ax.set_title(class_names[labels[i]], fontsize=7)
    ax.axis("off")
plt.suptitle("Sample training batch (augmented)", fontsize=13)
plt.tight_layout()
plt.savefig("sample_batch.png", dpi=100, bbox_inches="tight")
plt.show()
print(f"Batch tensor shape: {images.shape}  | dtype: {images.dtype}")
```

```python
# ── Export shared objects for reuse in later notebook
# To guarantee all three models are trained and evaluated on exactly the same data splits, we export shared_data.pkl
# All downstream notebooks (ResNet, EfficientNet, Classical) will import from here
import pickle

shared_data = {
    "train_paths" : train_paths,  "train_labels" : train_labels,
    "val_paths"   : val_paths,    "val_labels"   : val_labels,
    "test_paths"  : test_paths,   "test_labels"  : test_labels,
    "class_names" : class_names,
    "class_to_idx": class_to_idx,
    "class_weights_tensor": class_weights_tensor,
    "IMG_SIZE"    : IMG_SIZE,
    "SEED"        : SEED,
}

with open("shared_data.pkl", "wb") as f:
    pickle.dump(shared_data, f)

print("shared_data.pkl saved — import this in all subsequent notebooks.")
print("\n📋 Summary:")
print(f"   Classes     : {len(class_names)}")
print(f"   Train/Val/Test: {len(train_paths)} / {len(val_paths)} / {len(test_paths)}")
print(f"   Image size  : {IMG_SIZE}×{IMG_SIZE}")
print(f"   Batch size  : {BATCH_SIZE}")
```
